## Creating new dataframe

```python
# The statements below Transpose the wavelengeth data 
# Casts the Wavelenth column to integer type
# Adds a new column WavePosition

dfwave = wavelength_data.T
dfwave = dfwave[1:]
dfwave = dfwave.astype(int)
dfwave.columns = ['Wavelength']
dfwave['WavePosition'] = dfwave.index
dfwave.head()
```

```python
""" Plot the average spectra for each of the five vanilla products for the full range (900-1700nm)"""

make_plot(final_avg_spectra)
plt.xlim(900, 1700)
plt.ylim(0.3, 2.5)
plt.show()
```

```python
""" Plot the average spectra for each of the five vanilla products for the range (900-1350nm)"""

sub_data = final_avg_spectra[final_avg_spectra.Wavelength < 1350]
make_plot(sub_data)
plt.xlim(900, 1350)
plt.ylim(0.3, 0.5)
plt.show()
```

```python
""" Plot the average spectra for each of the five vanilla products for the range (1400-1500nm)"""

sub_data_upper = final_avg_spectra[(final_avg_spectra.Wavelength>1399) & (final_avg_spectra.Wavelength < 1501)]
make_plot(sub_data_upper)
plt.xlim(1400, 1500)
plt.ylim(1, 2.5)
plt.show()
```

# Part 2: Create an LDA Model and Plot in 2-Dimensions

Linear Discriminant Analysis models are useful with spectroscopy data, especially when doing classification problem, since they give a very quick indication of whether there is enough information contained in the spectra to discriminate between the various classes. In this part of the exercise, you are required to build an LDA model using Scikit_Learn's implementation of [Linear Discriminant Analysis](http://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html#sklearn.discriminant_analysis.LinearDiscriminantAnalysis) 

A common pre-processing technique in spectroscopy is to smooth the data, and also to use either or both the first and second derivatives of each wavelength. The following function performs both smoothing, as well as calculating the first and second derivatives. This function assumes that the first column in your dataframe has an identifier (e.g. 'Scan'), and that the last two columns in your dataframe are called 'Serial' and 'Target'. All the other columns are assumed to be spectral data. 

Set the parameters smooth_y, smooth_y1 , and smooth_y2 to 5,15,15 respectively. 

```python
from scipy.signal import savgol_filter
def expandSpectra(df,smooth_y,smooth_y1,smooth_y2): #funtion to smooth and expand NIR spectra by adding 1st and 2nd derivatives
    df.reset_index(drop=False)
    #extract the data (i.e. everything excluding the Scan ID and the SampleType)
    x = df.as_matrix(df.columns[0:df.shape[1]-2])
    #apply a SavGol filter to smooth raw data
    y = savgol_filter(x, smooth_y, 2, deriv=0)
    #apply a 1st derivative SavGol filter
    y1 = savgol_filter(x, smooth_y1, 2, deriv=1)
    #apply a 2nd derivative SavGol filter
    y2 = savgol_filter(x, smooth_y2, 2, deriv=2)
    #extract the column names
    cols = df.columns[0:df.shape[1]-2]
    #create dataframe for smoothed raw data
    dftarget=df['Target'].copy()
    dftarget=pd.Series.to_frame(dftarget)
    df0=pd.DataFrame(data=y, columns=cols)
    #append 1st and 2nd derivative filtered features to dataframe
    f_df = pd.DataFrame(data=y1, columns=cols)
    for col_name in f_df.columns:
        f_df.rename(columns={col_name: "1st-" + str(col_name)}, inplace=True)
    s_df = pd.DataFrame(data=y2, columns=cols)
    for col_name in s_df.columns:
        s_df.rename(columns={col_name: "2nd-" + str(col_name)}, inplace=True)
    #combine all three dataframes
    dfexpanded = pd.concat([df0, f_df], axis=1)
    dfexpanded = pd.concat([dfexpanded, s_df], axis=1)
    dftarget.reset_index(inplace=True)
    dfexpanded = pd.concat([dfexpanded, dftarget], axis=1)
    return dfexpanded;
```

```python
""" Apply the expandSpectra function to the dataframe that you created for the graphs. 
Set the parameters smooth_y, smooth_y1 , and smooth_y2 to 5,15,15 """

dfexpanded = expandSpectra(vanilla_spectral_data, 5, 15, 15)
dfexpanded.set_index('Scan',inplace=True)
dfexpanded.head()
```

## Standardize the Data
It is also common in machine learning to standardize features by removing the mean and scaling to unit variance. Once you have expanded the spectra, apply [Scikit-Learn's Standard Scaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) function. 

```python
""" Apply the Standard Scaler function to the dataframe that you created for the graphs """
from sklearn.preprocessing import StandardScaler

smooth_spectra = dfexpanded[dfexpanded.columns[0:dfexpanded.shape[1]-1]]
target = dfexpanded.Target

scaler = StandardScaler().fit(smooth_spectra)
scaled_spectra = scaler.transform(smooth_spectra)
```

## Create an LDA model
Create an LDA model using two components ("n_components=2"), and fit your standardised data to this.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
""" Create an LDA model on the standardized data """

lda = LinearDiscriminantAnalysis(n_components=2) #creating LDA instance
x_lda = lda.fit(scaled_spectra, target).transform(scaled_spectra) # fit the scaled data to the lda instance and transform it 
```

## Plot the Two Components of the LDA Transformed Data
Create a scatter plot of the two components. You should end up with a plot that looks like the following.

```python
from IPython.display import Image
Image(filename='./graphs/vanilla-LDA.png', width=650)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
""" Create an LDA model on the standardized data """

target_names = target.unique() # define target_names as the unique vanilla extracts
colors = ['green', 'red', 'brown', 'blue', 'yellow'] # specific colors to use for the chart

plt.figure(figsize=(10,8))

# the for loop goes through a collectioin of target_names and colors and used that for plotting
for color, target_name in zip(colors, target_names):
    plt.scatter(x_lda[target==target_name, 0], x_lda[target==target_name, 1], alpha=0.8, color=color,
                label=target_name)

plt.legend(loc='upper center', bbox_to_anchor=(0.5,-0.1))
plt.title('Linear Discriminant Analysis: NIR Absorbance of Vanilla Extracts')
plt.xlabel('LDA_1')
plt.ylabel('LDA_2')
plt.show()
```
